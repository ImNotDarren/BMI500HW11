```python
from environment import Environment
import matplotlib.pyplot as plt
```

### A)

```python
def run_env(p, q, grid_size = 75, agent_num = 100, agent_mode = 'random'):
    env = Environment(p=p, q=q, grid_size=grid_size, agent_num=agent_num, agent_mode=agent_mode)
    env.initialize()

    step = []
    susceptible = []
    infected = []
    recovered = []

    for i in range(200):
        step.append(i + 1)
        env.step()
        curr_s, curr_i, curr_r = env.count_infection()
        susceptible.append(curr_s)
        infected.append(curr_i)
        recovered.append(curr_r)

    plt.plot(step, susceptible, label='Susceptible')
    plt.plot(step, infected, label='Infected')
    plt.plot(step, recovered, label='Recovered')
    plt.xlabel('Step')
    plt.ylabel('Number of agents')
    plt.legend()
    plt.show()
```

```python
run_env(0.05, 0.02)
```

```python
run_env(0.1, 0.05)
```

```python
run_env(0.8, 0.01)
```

```python
run_env(0.1, 0.05, 30, 900)
```

```python
run_env(0.8, 0.05, 30, 900)
```
